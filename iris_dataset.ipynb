```python
# Select only those rows where:
# • petal_length > 4.5
# • species = "Iris-virginic
```

```python
filtered_df = df[
    (df["petal_length"] > 4.5) &
    (df["species"] == "Iris-virginica")
]

print(filtered_df)
```

```python
# Group by and compute:
# • Average sepal_length
# • Maximum petal_width
# • Standard deviation of sepal_width
```

```python
print(df.max(numeric_only=True))
```

```python
df["petal_width"].max()
```

```python
df["petal_width"].std()
```

```python
df["sepal_length"].mean()
```

```python
grouped = df.groupby("species")

result = grouped.agg(
    avg_sepal_length = ("sepal_length", "mean"),
    max_petal_width  = ("petal_width", "max"),
    std_sepal_width  = ("sepal_width", "std")
)

print(result)
```

```python
df
```

```python
# Create a new column:
# "petal_ratio" = petal_length / petal_width
# • For each species, find the average petal_ratio.

df["petal_ratio"] = df["petal_length"] / df["petal_width"]
df
```

```python
group = df.groupby("species")

avg_petal_ration = group.agg(
    average = ("petal_ratio","mean")
)
print(avg_petal_ration)
```

```python
titanic = pd.read_csv("Titanic-Dataset.csv")
```

```python
titanic[["Name","Sex","Age","Fare","Survived"]] # multiple kae loya double square bracket
```

```python
# Select passengers who:
# • Are female
# • Fare > 30

select_passengers = titanic[
    (titanic["Sex"] == "female") &
    (titanic["Fare"] > 30)
]
select_passengers.head()
```

```python
# Group by  pclass and compute:
# • Survival rate
# • Average Fare
# • Average Age

grouped = titanic.groupby("Pclass")

result = grouped.agg(
    survival_rate  = ("Survived","mean"),
    average_fare = ("Fare","mean"),
    average_age = ("Age","mean")
)
result
```
