```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
```

```python
# plot data now 
plt.plot(years,oscar_revenue)
plt.title("Oscar Movie Revenue In Each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
```

```python
plt.plot(years,oscar_revenue,"o--r",label = "Oscar Movies")
plt.plot(years,non_oscar_revenue , label = "Non Oscar Movies")


plt.title("Oscar movie Vs Non Oscar Movie Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")

plt.legend()
# plt.legend(loc = "upper left")

# plt.style.available

# plt.style.use("dark_background")
plt.style.use("ggplot")
# plt.grid()
# plt.style.use("xkcd")
```
