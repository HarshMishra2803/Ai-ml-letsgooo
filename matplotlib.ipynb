```python
# Real data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]

oscar_revenue = [1005, 170, 427, 133, 232]  # in $M
```

```python
# plot data now 
plt.plot(years,oscar_revenue)
plt.title("Oscar Movie Revenue In Each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()
```

```python
# Non-Oscar Winner Movies
non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
```

```python
import numpy as np;
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = [
    "Slumdog Millionaire",
    "Avatar",
    "Inception",
    "Hugo",
    "Lincoln"
]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M
years = [2008, 2009, 2010, 2011, 2012]


# we customize width . width control how thick each bar is , usually ranges from 0 to 1 .[default value is 0.8]

# 2 bar charts sath mae

x= np.arange(len(years))
width = 0.4

plt.title("Oscar Vs Non Oscar Movies Revenue")
plt.xlabel("Years")
plt.ylabel("Revenue($M)")
plt.bar(x-width/2,oscar_revenue,width,label="oscar_movies")
plt.bar(x+width/2,non_oscar_revenue,width ,label = "non_oscar_movies")


plt.xticks(x,years)

plt.legend()
```

```python
plt.barh(oscar_movies,oscar_revenue)
plt.xlabel("Revenue")
plt.ylabel("Oscar_movies")
plt.title("Details about Revenue per OscarMovies")
```

```python
# multiple datasets

legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]


# trx amount distrubution - legit vs fraud
```

```python
plt.hist(legit_transactions , color ="green" ,bins = 20 ,label = "legit",edgecolor = "black",alpha=0.5)
plt.hist(fraud_transactions , color ="red" ,bins = 20 ,label = "fraud",edgecolor = "black",alpha = 0.5)

plt.title("trx amount distrubution - legit vs fraud")
plt.xlabel("Amount(in $)")
plt.ylabel("Frequency")
plt.grid(True)
plt.legend()
```
