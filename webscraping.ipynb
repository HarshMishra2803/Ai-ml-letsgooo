```python
#  WEB SCRAPING 

import requests
```

```python
URL = "https://www.scrapethissite.com/pages/simple/"

res = requests.get(URL)
```

```python
res
```

```python
print(res.status_code)
```

```python
if res.status_code == 200 :
    print(res.headers)  # res.content , res.text 
```

```python
# downloading and storing this data

with open("scraped_data/data1.html","w") as f:
    f.write(res.text)
```

```python
 # beautiful soup

from bs4 import BeautifulSoup
```

```python
with open("scraped_data/data1.html","r") as f :
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content, "lxml")

```

```python
# soup 
```

```python
# methods
soup.find("h1")
soup.find("h3")
soup.h1
```

```python
all_h3 = soup.find_all("h3")
```

```python
type(all_h3)
```

```python
for h3 in all_h3:
    # print(h3)
    print(h3.get_text(strip=True))
    print(h3.find_next("div").select("span.country-population")[0].get_text(strip = True))
```
