```python
#ques1 - select only rows where petal_length > 4.5 , species = " Iris-virginica"

select_row = iris [(iris["petal_length"] > 6.5 ) & (iris["species"] ==  "Iris-virginica") ]

```

```python
print(select_row)
```

```python
# group by and compute 
# avergae sepal_ length , maximum petal lenght , standard deviation of sepal_width 

grouped = iris.groupby("species")
```

```python
result = grouped.agg(
    avg_sepal_length = ("sepal_length","mean"),
    max_petal_length = ("petal_length","max"),
    std_sepal_width = ("sepal_width","std"))
    
```

```python
print(result)
```

```python
iris["petal_ratio"] = iris["petal_length"]/iris["petal_width"]
group = iris.groupby("species")

avg_petal_ratio = group.agg(
    average = ("petal_ratio","mean")
)
```

```python
print(avg_petal_ratio)
```
