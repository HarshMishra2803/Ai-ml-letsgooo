```python
#execution performance
size = 1_000_000

#python list
py_list = list(range(size))
start = time.time()
sq_list =[ x**2 for x in py_list]
end = time.time()
print(f"time taken to complete the execution {end-start} seconds")

#numpy array
py_array = np.array(py_list)
start =time.time()
sq_array = py_array**2
end =time.time()
print(f"time taken to complete the execution {end-start} seconds")

```
