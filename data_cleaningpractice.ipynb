```python
import json
```

```python
#load the data
def data_load(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data    
    
```

```python
data=data_load("user_data.json")
```

```python
print(data,type(data))
```

```python
#data cleaning -- rating 
```

```python
def rate_cleaner(data):
    text_to_num = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5
    }

    for user in data:
        raw_rating = user["rating"]

        if isinstance(raw_rating, str):
            raw_rating = raw_rating.strip().lower()

            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]
            else:
                raw_rating = float(raw_rating)

        user["rating"] = raw_rating


    return data
```

```python
data = rate_cleaner(data)
```

```python
print([user["rating"] for user in data])
```

```python
def age_cleaner(data):
    for user in data:
        raw_age = user["age"]

        if isinstance(raw_age, str):
            raw_age = int(raw_age)
        elif raw_age is None:
            raw_age = None

        user["age"] = raw_age

    return data
```

```python
data = age_cleaner(data)
```

```python
print([users["age"] for users in data])
```

```python
def duplicate_name_cleaner(data):
    clean_data = []
    unique_user = set()

    for user in data:
        name = user["name"].strip()

        if name in unique_user:
            continue

        unique_user.add(name)
        clean_data.append(user)

    return clean_data
```

```python
print([users["name"] for users in data])
```

```python
data = duplicate_name_cleaner(data)
```

```python
data = duplicate_name_cleaner(data)

for user in data:
    print(user["name"])
```
