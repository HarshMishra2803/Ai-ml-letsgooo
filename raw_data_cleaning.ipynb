```python
import pandas as pd
# data cleaning  -- 1 null vlaue , 2 deduplication , 3 difffernt types of data - transformation 

# ____ HAndle missing Value _____
```

```python
# 1 - missing values isnull()

df = pd.read_csv("raw_data.csv")

df.isnull()  # jaha pae null value ha ya Nan ha 
df.isna() # same kamm ha isnull jasa

# different function kee chaining
df.isnull().sum()

# dropna() -- null vlaue ko drop kardetaha row ko pura # rows
df.dropna()

# column ko drop karna ha 
df.dropna(axis=1) # column ko drop karta ha 


# fillna -- jo null vlaue usko kisi particular value sae fill karsakta ha
df.fillna(0)

age_mean = df["age"].mean() # avergae value  # lakin ek chiz original mae koi chnage nai hoga
cleaned_data = df.copy()
cleaned_data["age"]=cleaned_data["age"].fillna(age_mean)
cleaned_data

# ffill and bfill - ffill stands for forward fill -- upar sae nicha kae forward kardega upar kee vlaue ko neecha copy kardega asa he backward fill


df.ffill()
df.bfill()
```
