```python
import json
```

```python
#load the data
def load_data(filename):
    with open (filename,"r") as f:
        data = json.load(f)
        
    return data    
```

```python
data =load_data("store_data.json")
```

```python
print(data,type(data))
```

```python
def clean_data(data):
    text_to_num = {"one":1 ,"two":2 , "three":3, "four":4 , "five":5}
    cleaned_data = []
    unique_user = set()

    for user in data:
        raw_rating = user["rating"]

        if isinstance(raw_rating, str):
            raw_rating = raw_rating.strip().lower()
            if raw_rating in text_to_num:
                raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        if user.get("age") is None:
            user["age"] = None

        name = user["name"].strip()
        if name in unique_user:
            continue

        unique_user.add(name)
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)
```
