```python
 # melt and pivot
# melt -- wide to long format 
# pivot -- long to wide format

df = pd.DataFrame({
    "country": ["USA","USA","INDIA","INDIA"],
    "year": [2020,2021,2020,2021],
    "sales": [100,120,90,110],
    "profit": [20,25,15,18]
})

melted_df = df.melt(
    id_vars=["country", "year"],
    value_vars=["sales", "profit"],
    var_name="metrics",
    value_name="value"
)
melted_df
```

```python
original = melted_df.pivot(
    index = ["country" ,"year"],
    columns = "metrics",
    values = "value"
)
original
```

```python
df_customers = pd.DataFrame({
    "customer_id":[101,102,103,104],
    "name" : ["harsh","harsh1", "harsh2","harsh3"]
})

df_orders = pd.DataFrame({
    "order_id":[101,102,103,104],
     "customer_id":[2,1,4,5],
     "amount":[250,120,300,180]
    
})

pd.merge(df_customers ,df_orders ,on="customer_id") # innere join 

# left join 

pd.merge(df_customers ,df_orders ,on="customer_id",how="left")
pd.merge(df_customers ,df_orders ,on="customer_id",how="right") # right join
pd.merge(df_customers ,df_orders ,on="customer_id",how="outer") # outer join
```
